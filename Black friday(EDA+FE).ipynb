# Problem statement: 
A retail company "ABC private limited" wants to understand the customer purchase behaviour(specifically, purchase amount) against various products of diff categories. They have shared the purchase summary of various customers for selected high vol products from last month. It also contains customers demographics , product details and total purchase amt of last month.

Now, they want to build a model to predict the purchase amt of customer against the products which will help them to create personalised offer for customers against diff products.
