```python
sns.heatmap(df.isnull(),yticklabels=False, cbar=False, cmap='viridis')
# df.isnull() returns a DataFrame of the same shape, where each cell contains True if the value is NaN (missing)
# yticklabels= This argument disables the y-axis labels, so the heatmap won't display row names or labels
# cbar= This argument removes the color bar that typically appears alongside the heatmap
# cmap = This specifies the color map to use for the heatmap. viridis is a popular color map that goes from dark purple to bright yellow. 
```

```python
df_country=pd.read_excel('Country-code.xlsx')
df_country.head()
```

```python
df.shape
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
# 'on' is used to chcek that on which column you are going to combine-left or right.
```

```python
final_df.head(2)
```

```python
# To check Data types:
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names=final_df.Country.value_counts().index
# to find no. of country and records associated with it
```

```python
country_val=final_df.Country.value_counts().values
```

```python
country_val
```

```python
#  Pie Chart- Top 3 country's that uses zomato:

plt.pie(country_val[:3], labels=country_names[:3],autopct="%1.2f%%");
# %1.2f%% = format - after decimal 2 places will be visible.
```

Observation: Zomato max records/ transactions are from India
after then USA and UK

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
# reset_index() = used to convert into df
```

```python
ratings
```

## Observation:
* When rating is from(4.5,4.9) =Excellent
* When rating is from(4.0,4.4) = Very good
* When rating is from(3.5,3.9) = good
* When rating is from(2.5,3.4) = average
* When rating is from(1.8,2.4) = poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating", y="Rating Count", data= ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count", hue='Rating color', palette=['white', 'red','orange','yellow','green','green'], data= ratings)
# to make colors according to given data, we use palette function.
sns.set_style("darkgrid")
```

## Observation:
* Not rated count is very high.
* max ratings are between 2.5- 3.4

```python
## Count Plot - Used for plotting according to categorial variables.

sns.countplot(x="Rating color", data=ratings, legend=False, palette=['white', 'red','orange','yellow','green','green'])
```
